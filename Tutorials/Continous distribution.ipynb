```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
a1 = np.random.normal(0,1,1000)
x1 = np.sort(a1)
```

```python
px1 = (1/(1*np.sqrt(2*math.pi)))*np.exp(-0.5 * (x1-0)**2)
```

```python
a2 = np.random.normal(2,2,1000)
x2 = np.sort(a2)
```

```python
px2 = (1/(2*np.sqrt(2*math.pi)))*np.exp(-0.25 * (x2-2)**2)
```

```python
a3 = np.random.normal(0,2,5000)
x3 = np.sort(a3)
px3 = (1/(2*np.sqrt(2*math.pi)))*np.exp(-0.25 * (x3-0)**2)
```

```python
b1 = np.random.exponential(0.5,5000)
X1 = np.sort(b1)
b2 = np.random.exponential(1,5000)
X2 = np.sort(b2)
b3 = np.random.exponential(1.5,5000)
X3 = np.sort(b3)
```

```python
Y1 = 0.5 * (np.exp(-0.5*X1))
Y2 = 1 * (np.exp(-1*X2))
Y3 = 1.5 * (np.exp(-1.5*X3))
```

```python
plt.style.use('default')
plt.figure(figsize=(16, 6))
plt.subplot(1,2,1)

plt.plot(x1,px1 , c= "r",linewidth=3 , label = " =0 &  =1 ")
plt.plot(x2,px2 , c= "b",linewidth=3 , label = " =2 &  =2 ")
plt.plot(x3,px3 , c= "g",linewidth=3 , label = " =0 &  =2 ")
plt.title("Normal or Gaussian distribution",fontsize=20)
plt.ylabel("probability density f(X)", fontsize=10)
plt.xlabel("X",fontsize = 10)
plt.xlim(-8,8)
plt.axvspan(-8,8,color='White')
plt.grid()
plt.legend()
plt.subplot(1,2,2)
plt.plot(X1,Y1 , c= "g",linewidth=3 , label = " = 0.5 ")
plt.plot(X2,Y2 , c= "r",linewidth=3 , label = " = 1 ")
plt.plot(X3,Y3 , c= "b",linewidth=3 , label = " = 1.5")
plt.title("Exponential distribution" , fontsize =20)
plt.ylabel("probability density f(X)", fontsize=10)
plt.xlabel("X",fontsize = 10)
plt.xlim(0,5)
plt.axvspan(0,5,color='White')
plt.legend()
plt.grid()
```
